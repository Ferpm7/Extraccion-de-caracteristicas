#Debido a que hay columnas con muy pocos datos y estos no se pueden imputar ya que pertenecen o son datos que son obtenidos de personas como clientes etc. No hay posibilidad de llenar los datos a otras columnas pues sería información erronea al asignar información de una persona a otra persona como es el caso de whatsapp, ya que como se puede apreciar solo contiene 42 datos y para el caso de e-mail solo 29, por tanto, solo utilizarenmos los datos que contengan al menos el 60% de los datos por columna

```python
porcentaje_nulos = df.isnull().mean()
umbral = 60  # Umbral del 76%
columnas_a_eliminar = porcentaje_nulos[porcentaje_nulos >= umbral / 100].index.tolist()
df = df.drop(columns=columnas_a_eliminar)
df
```

#Aplicar métodos para reemplazar valores nulos del dataframe

```python
#Identificar valores nulos por columna
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
#Identificar valores nulos por dataframe
valores_nulos=df.isnull().sum().sum()
valores_nulos
```

#Dividimos en 2 el dataframe original para trabajar de mejor manera con 2 dataframes, 1 con puras variables categoricas y otro con puras variables numericas. De tal modo que nos permita trabajar de manera más facil y rapida la limpieza.

```python
# Obtén las columnas numéricas
df_numeric = df.select_dtypes(include=['int64', 'float64'])

# Obtén las columnas categóricas
df_categoric = df.select_dtypes(include=['object'])

# Ahora 'df_numeric' contiene solo las columnas numéricas y 'df_categoric' contiene solo las columnas categóricas
```

```python
df_numeric
```

```python
# Itera a través de las columnas del DataFrame
for columna in df_numeric.columns:
    if df_numeric[columna].isnull().any():
        # Verifica si la columna tiene valores nulos
        if pd.api.types.is_numeric_dtype(df_numeric[columna]):  # Verifica si la columna es numérica
            if pd.api.types.is_integer_dtype(df_numeric[columna]):  # Verifica si la columna es de tipo int64
                promedio = df_numeric[columna].mean()
                df_numeric[columna] = df_numeric[columna].fillna(round(promedio))
            else:
                promedio = df_numeric[columna].mean()
                df_numeric[columna] = df_numeric[columna].fillna(round(promedio, 1))
```

```python
#Identificar valores nulos por columna
valores_nulos=df_numeric.isnull().sum()
valores_nulos
```

```python
#Identificar valores nulos por columna
valores_nulos=df_categoric.isnull().sum()
valores_nulos
```

```python
# Itera a través de las columnas del DataFrame
for columna in df_categoric.columns:
    if df_categoric[columna].isnull().any():
        # Verifica si la columna tiene valores nulos
        moda = df_categoric[columna].mode().iloc[0]  # Obtiene el valor más frecuente
        df_categoric[columna] = df_categoric[columna].fillna(moda)
```

#Aplicar métodos (Desviación Estándar o Rango Intercuartílico) para eliminar Outliers del dataframe

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
df_numeric.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=df_numeric
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
df_numeric= df_numeric[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df_numeric
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=df_numeric.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=df_numeric.copy()
data_clean_num=data_clean.fillna(round(df_numeric.mean(),1))
data_clean_num
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean_num.isnull().sum()
valores_nulos
```

```python
#Obtengo un análisis univariado de las variables categóricas
freq_tbl(Base_limpia)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from docx import Document
from docx.shared import Inches
import random

# Lista de nombres de las 20 variables categóricas
variables_categoricas = [
    '214_customer_relationship_tools',
    '232_type_of_store',
    '186_internet_connection',
    '108_does_the_micro_retailer_has_a_barred_window_',
    '5_change_store_space_last_year',
    '6_change_employees_average_salary_last_year',
    '99_does_the_micro_retailer_exhibits_products_outside_',
    '18_sales_records',
    '210_sales_channels',
    '189_payment_methods',
    '193_sales_planning_tools',
    '60_services',
    '_title',
    '_updated_by',
    '42_willingness_share_inventory_data_suppliers',
    '161_actions_stockouts',
    '157_frequency_organize_shelves',
    '185_place_orders_suppliers',
    '56_procurement_changes_pandemic',
    '267_technology_investment_customer_preferences'
]

# Diccionario que asocia una variable categórica con un tipo de gráfica al azar
graficas = {
    '214_customer_relationship_tools': 'bar',
    '232_type_of_store': 'pie',
    '186_internet_connection': 'pie',
    '108_does_the_micro_retailer_has_a_barred_window_': 'bar',
    '5_change_store_space_last_year': 'pie',
    '6_change_employees_average_salary_last_year': 'pie',
    '99_does_the_micro_retailer_exhibits_products_outside_': 'pie',
    '18_sales_records': 'pie',
    '210_sales_channels': 'bar',
    '189_payment_methods': 'pie',
    '193_sales_planning_tools': 'bar',
    '60_services': 'bar',
    '_title': 'bar',
    '_updated_by': 'pie',
    '42_willingness_share_inventory_data_suppliers': 'pie',
    '161_actions_stockouts': 'bar',
    '157_frequency_organize_shelves': 'line',
    '185_place_orders_suppliers': 'bar',
    '56_procurement_changes_pandemic': 'bar',
    '267_technology_investment_customer_preferences': 'pie'
}

# Crear un documento Word
doc = Document()

# Itera a través de las variables categóricas
for variable in variables_categoricas:
    # Obtén un análisis univariado de la variable categórica en específico
    table1 = freq_tbl(Base_limpia[variable])
    print(f'Table1 para {variable}:\n{table1}\n')

    # Obtén un análisis univariado de la variable categórica en específico
    table = Base_limpia[variable].value_counts().reset_index()
    table.columns = [variable, 'frequency']

    # Filtra los valores más relevantes de la variable categórica seleccionada
    filtro = table[table['frequency'] > 1]

    # Ajusta el índice del DataFrame filtrado
    filtro_index = filtro.set_index(variable)

    # Imprimir filtro y filtro_index
    print(f'Filtro para {variable}:\n{filtro}\n')
    print(f'Filtro Index para {variable}:\n{filtro_index}\n')

    # Selecciona un tipo de gráfica al azar del diccionario
    tipo_grafica = graficas[variable]

    # Realiza la gráfica correspondiente al tipo seleccionado
    plt.figure(figsize=(10, 7))  # Tamaño ajustado para todas las gráficas

    if tipo_grafica == 'bar':
        filtro_index.plot(kind='bar', width=1)
        plt.title(f'{variable} - Gráfico de barras')
    elif tipo_grafica == 'pie':
        filtro_index['frequency'].plot(kind='pie', autopct='%1.1f%%')
        plt.title(f'{variable} - Gráfico de pastel')
    elif tipo_grafica == 'scatter':
        filtro_index.plot('frequency', 'frequency', kind='scatter')
        plt.title(f'{variable} - Gráfico de dispersión')
    elif tipo_grafica == 'area':
        filtro_index.plot(kind='area', alpha=0.8)
        plt.title(f'{variable} - Gráfico de área')
    elif tipo_grafica == 'line':
        filtro_index.plot(kind='line')
        plt.title(f'{variable} - Gráfico de líneas')

    plt.xlabel('Valores')
    plt.ylabel('Frecuencia')

    # Guardar la gráfica como una imagen
    plt.savefig(f'grafica_{variable}.png')

    # Agregar la gráfica al documento Word con un tamaño personalizado
    doc.add_picture(f'grafica_{variable}.png', width=Inches(6.0), height=Inches(3.5))

    plt.tight_layout()
    plt.show()

# Guardar el documento Word
doc.save('REPORTE DE RESULTADOS.docx')
```
